```python
import pandas as pd

# 1. Cargamos el dataset inmobiliario
df_housing = pd.read_csv('Nashville Housing sucio.csv')

# 2. Hacemos un reporte rápido de cuántos valores nulos (vacíos) hay en cada columna
print("🚨 REPORTE DE VALORES NULOS 🚨")
print(df_housing.isnull().sum())

# 3. Miramos las primeras filas para entender la estructura
df_housing[['ParcelID', 'PropertyAddress', 'SalePrice', 'OwnerName']].head()
```

```python
# 1. SALVATAJE DE NULOS: Rellenamos los vacíos sin borrar las filas
df_housing['OwnerName'] = df_housing['OwnerName'].fillna('Sin Datos')

# 2. PARSING DE DIRECCIONES: Separamos la dirección usando la coma como machete
direcciones = df_housing['PropertyAddress'].str.split(',', n=1, expand=True)

# 3. Guardamos las nuevas columnas limpiando los espacios extra (strip)
df_housing['Address'] = direcciones[0].str.strip()
df_housing['City'] = direcciones[1].str.strip()

# 4. Verificamos la magia
df_housing[['PropertyAddress', 'Address', 'City', 'OwnerName']].head(10)
```

```python
# 1. ESTANDARIZACIÓN BOOLEANA: Usamos un diccionario para reemplazar 'Y' y 'N'
df_housing['SoldAsVacant'] = df_housing['SoldAsVacant'].replace({'Y': 'Yes', 'N': 'No'})

# 2. NORMALIZACIÓN TEMPORAL: Convertimos el texto a fecha internacional (YYYY-MM-DD)
df_housing['SaleDate'] = pd.to_datetime(df_housing['SaleDate']).dt.strftime('%Y-%m-%d')

# 3. Auditoría rápida: le pedimos a Python que nos muestre los valores únicos que quedaron
print("Valores únicos en SoldAsVacant ahora:")
print(df_housing['SoldAsVacant'].unique())

# 4. Vemos cómo quedó la tabla de fechas
df_housing[['SaleDate', 'SoldAsVacant']].head()
```

```python
# Guardamos nuestra obra maestra del Real Estate
df_housing.to_csv('03_Nashville_Limpio_Python.csv', index=False)
```
