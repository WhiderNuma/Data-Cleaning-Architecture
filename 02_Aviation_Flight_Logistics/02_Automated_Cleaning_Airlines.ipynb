```python
import pandas as pd

# 1. Cargamos el dataset sucio de vuelos
df_vuelos = pd.read_csv('Airline Dataset Updated - v2 sucio.csv')

# 2. Vamos a mirar específicamente cómo están rotas las fechas
print(df_vuelos['Departure Date'].head(15))
```

```python
# 1. Le pedimos a Pandas que use su inteligencia para leer el caos y transformarlo en una Fecha Real
# El parámetro errors='coerce' hace que si hay un texto totalmente ilegible, lo deje vacío (NaT) en vez de romper todo el código.
df_vuelos['Departure Date'] = pd.to_datetime(df_vuelos['Departure Date'], errors='coerce')

# 2. Una vez que Python entiende que son fechas, le damos el formato corporativo internacional (YYYY-MM-DD)
df_vuelos['Departure Date'] = df_vuelos['Departure Date'].dt.strftime('%Y-%m-%d')

# 3. Vemos el resultado
df_vuelos[['Passenger ID', 'First Name', 'Departure Date']].head(10)
```

```python
# 1. Contamos cuántas filas tenemos antes de la limpieza
total_antes = len(df_vuelos)

# 2. Eliminamos los duplicados usando el ID único del pasajero como referencia
# keep='first' le dice a Python que se quede con la primera aparición y borre los clones
df_vuelos = df_vuelos.drop_duplicates(subset=['Passenger ID'], keep='first')

# 3. Contamos cuántas filas sobrevivieron
total_despues = len(df_vuelos)
eliminados = total_antes - total_despues

# 4. Mostramos el reporte de impacto
print(f"Total de registros originales: {total_antes}")
print(f"Total de registros limpios: {total_despues}")
print(f"¡Se eliminaron {eliminados} registros duplicados que inflaban las métricas!")
```

```python
# 1. Strip: Cortamos cualquier espacio en blanco invisible al principio o al final del texto
df_vuelos['Airport Name'] = df_vuelos['Airport Name'].str.strip()
df_vuelos['Arrival Airport'] = df_vuelos['Arrival Airport'].str.strip()

# 2. Upper: Forzamos a que todos los códigos de aeropuerto estén en MAYÚSCULAS
df_vuelos['Arrival Airport'] = df_vuelos['Arrival Airport'].str.upper()

# 3. Title: Normalizamos el estado del vuelo para que la primera letra sea mayúscula (Ej: 'On Time')
df_vuelos['Flight Status'] = df_vuelos['Flight Status'].str.strip().str.title()

# 4. Comprobamos la limpieza de texto
df_vuelos[['Airport Name', 'Arrival Airport', 'Flight Status']].head(10)
```

```python
# Guardamos la tabla limpia sin el índice de Pandas
df_vuelos.to_csv('02_Airline_Limpio_Python.csv', index=False)
```
